```python
import pandas as pd 
df = pd.read_excel('https://github.com/datagy/pivot_table_pandas/raw/master/sample_pivot.xlsx', parse_dates=['Date'])
df.head()
```

```python
sales_by_region = pd.pivot_table(df, index = 'Region')
sales_by_region
```

```python
sales_by_region.plot(kind='bar')
```

```python
#Multiple Indexes
sales_by_region = pd.pivot_table(df, index = ['Region','Type'])
sales_by_region
```

```python
sales_by_region = pd.pivot_table(df, index = ['Region','Type'] , values='Sales')
```

```python
sales_by_region.plot(kind='bar')
```

```python
df1 = df.loc[:300,:]
df1
```

```python
df2 = df.loc[301:700,:]
df2
```

```python
df3 = df.loc[701:1000,:]
df3
```

```python
df_concat = pd.concat([df1,df2,df3], keys=['df1','df2','df3'])
df_concat
```
